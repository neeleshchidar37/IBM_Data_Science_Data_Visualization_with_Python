### Introduction to Matplotlib and Line Plots

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv")
```

```python
df.head()
```

```python
# set index to Country
df.set_index('Country', inplace=True)
df.head(2)
```

```python
# remove index name
df.index.name = None
df.head(2)
```

```python
# Convert years to string
years = list(map(str, range(1980, 2014)))
years
```

```python
# Visualizing data using matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print('Matlplotlib version: ', mpl.__version__)
```

```python
print(plt.style.available)
mpl.style.use(['ggplot']) # for ggplot like style
```

```python
# plot a line graph of immigration from Haiti using df.plot()

years = list(map(str, range(1980, 2014)))
haiti = df.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot() # simple line plot
```

```python
# lets label the x and y axis
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.show()
```

```python
# annotate the 2010 Earthquake
haiti.index = haiti.index.map(int)

haiti.plot(kind='line')
plt.title('Immgration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

```python
# Lets compare the number of immigrants from Indian and China from 1980-2013

df_CI = df.loc[['India', 'China'], years]
df_CI.head()
```

```python
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind='line')
plt.title('Immigrants from India and China')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
# Compare the trends of top5 countries that contributed the most to immigration to Canada

df.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df.head(5).copy() # to avoid pandas dataframe warning
df_top5.drop(columns=['Continent', 'DevName', 'Region', 'Total'], inplace=True)
df_top5 = df_top5.transpose()

df_top5.plot(kind='line', figsize=(14,8))
plt.title('Immigration trend of top5 countries')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.legend(title='Country')
plt.show()
```
