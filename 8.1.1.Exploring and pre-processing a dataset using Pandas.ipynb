### Exploring and pre-processing a dataset using Pandas

##### The Dataset: Immigration to Canada from 1980-2013
##### Dataset Source: international migration flows to and from selected countries - The 2015 revision

```python
# the dataset contains annual data on the flows of international immigrants as recorded by the countries of destination.
# the data presents both inflows and outflows according to the place of birth, citizenship or place of previous/next
# residence both for foreigners and nationals.
# The current version presents data pertaining to 45 countries

# In this lab, we will focus on the Canadian immigration data.
```

```python
import pandas as pd
import numpy as np
```

```python
# read excel file , sheet = 'Canada by Citizenship'
# skipping first 20 rows and last 2 rows

df = pd.read_excel('canada.xlsx', sheet_name = 'Canada by Citizenship' , skiprows=range(20), skipfooter=2)

print('data read into a pandas dataframe')
```

```python
# in pandas axis=0 represnts rows and (default) and axis=1 represnts columns
# removing unnecessary columns
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df.head(2)
```

```python
df.head(2)
```

```python
# we can use rename() method by passing in a dictonary of old and new names as follows:

df.rename(columns={'OdName': 'Country', 'AreaName': 'Continent', 'RegName': 'Region'}, inplace=True)
df.columns
```

```python
# we will also add a 'Total' column that sums up the total immigrants by country over the entire period 1980-2013,
# Assuming the columns containing immigration data are from 1980 to 2013
# This code uses the iloc function to select columns from index 5 onwards (columns representing years 1980 to 2013),
# and then sum(axis=1) is used to calculate the sum of immigrants for each row

df['Total'] = df.iloc[:,4:].sum(axis=1)
```

```python
# filtering data based on crideria
# lets filter the dataframe to show the data on Asian countries (AreaName = Asia)

# 1. create the condition boolean series
condition = df['Continent'] == 'Asia'
condition
```

```python
# 2 pass this condition into the dataframe
df[condition]
```

```python
# we can pass multiple criteria in the same line
df[(df['Continent'] == 'Asia') & (df['Region'] == 'Southern Asia')]
```

```python
# Fetch the data where AreaName is 'Africa' and RegName is 'Southern Africa'

df[(df['Continent'] == 'Africa') & (df['Region'] == 'Southern Africa')]
```

```python
# Soring Values of a Dataframe or Series
# you can use the sort_values() function
# df.sort_values(by=['col_name'], axis=0, ascending=True, inplace=True/False, ignore_index=True/False)

df.sort_values(by='Total', axis=0, ascending=False, inplace=True)
top5 = df.head(5)
top5
```

```python
# Find out top 3 countries that contribures the most to immigration to Canada in the year of 2010
df_n = df
df_n.sort_values(by='2010', axis=0, ascending=False, inplace=True)
top_3 = df_n.head(3)
top_3
```
